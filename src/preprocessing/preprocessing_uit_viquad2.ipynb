```python
from pathlib import Path
import pandas as pd

# Find project root by walking up until data/uit_viquad2/raw exists.
root_dir = None
for candidate in [Path.cwd(), *Path.cwd().parents]:
    if (candidate / "data" / "uit_viquad2" / "raw").exists():
        root_dir = candidate
        break

if root_dir is None:
    raise FileNotFoundError("Không tìm thấy thư mục data/uit_viquad2/raw từ thư mục hiện tại")

train_path = root_dir / "data" / "uit_viquad2" / "raw" / "train-00000-of-00001.parquet"
df = pd.read_parquet(train_path)

print("Project root:", root_dir)
print("Rows:", len(df))
print("Columns:", list(df.columns))
display(df[["id", "question", "answers", "is_impossible"]].head())
```
